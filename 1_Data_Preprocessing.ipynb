# Outlier treatment
-Capping and flooring(the one we gonna use
-Exponential smoothing
-Sigma approach

```python
df.info()
```

```python
np.percentile(df.n_hot_rooms,[99])#taking 99th percentile of n_hot_rooms-its an array so we take[0] to make it one value
```

```python
np.percentile(df.n_hot_rooms,[99])[0]
```

```python
uv=np.percentile(df.n_hot_rooms,[99])[0]#upper value is set to differentiate 99th percentile value taken
```

```python
df[(df.n_hot_rooms>uv)]#shows the data where hotelroom number greater than uv
```

```python
df[df.n_hot_rooms>3*uv]=3*uv #values more than thrice upper value set to thrice of upper value
```

```python
sns.jointplot(x='n_hot_rooms',y='price',data=df) 
#compare previous plot with here,there max value was 90+ whereas here its 46 and the data is not as disturbed 
#because of the outliers as previously->Outliers treated using upper value logic!
```

```python
#Lets treat the outliers in rainfall using the lower value logic
#since in rainfall there is an outlier in the lower end
```

```python
lv=np.percentile(df.rainfall,[1])[0] 
#taken lower value as the 1th percentile value
```

```python
df[(df.rainfall<0.3*lv)]=0.3*lv
```

```python
df.describe()
#Now note the median(50th percentile) and mean of rainfall and n_hot_rooms are closer
```

# Missing values imputation

-Done according to what variable they are(categorical/quantitative)

-Usually filled up with mean,median,mode or random business specific way

```python
df.info()
#notice there are 8 missing values in n_hos_beds
```

```python
df.n_hos_beds=df.n_hos_beds.fillna(df.n_hos_beds.mean())
#fills missing data with mean
```

```python
df.info()
#now look there are all data cells filled
```
