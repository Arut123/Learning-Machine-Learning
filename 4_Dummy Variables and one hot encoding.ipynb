```python
import pandas as pd
df=pd.read_csv('carmodel.csv')
df
```

```python
dummies=pd.get_dummies(df.Car_Model)
dummies
```

```python
merged=pd.concat([df,dummies],axis='columns') #merges columns
merged
```

```python
#now you can drop one of the variables(here we also want to get rid of nan column)
```

```python
final=merged.drop(['Audi A5','Unnamed: 4'],axis='columns')
final
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
X=final.drop(['Sell_Price','Car_Model'],axis='columns')
X
```

```python
y=final.Sell_Price
y
```

```python
model.fit(X,y)
```

```python
model.score(X,y) #94 percent accurate!
```

```python
model.predict([[45000,4,0,0]]) #mileage 45000 4 yr old Audi A5 price
```

```python
model.predict([[86000,7,0,1]]) #mileage 86000 7 yr old Mercedes benz price
```
